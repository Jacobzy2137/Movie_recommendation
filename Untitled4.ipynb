popularity based movie recommendation system

```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_data = pd.read_csv('movies.csv')
movies_data.head()
```

```python
movies_data.shape
```

```python
selected_features = ['genres', 'keywords', 'tagline', 'director', 'cast']
```

```python
for feature in selected_features:
  movies_data[feature] = movies_data[feature].fillna('')
```

```python
combined_features = movies_data['genres'] + ' ' + movies_data['keywords'] + ' ' + movies_data['tagline'] + ' ' + movies_data['director'] + ' ' + movies_data['director']
```

```python
print(combined_features)
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

```python
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
movie_input = input('Enter movie name: ')
```

```python
list_of_titles = movies_data['title'].tolist()
print(list_of_titles)
```

```python
find_match = difflib.get_close_matches(movie_input, list_of_titles)
print(find_match)
```

```python
close_match = find_match[0]
print(close_match)
```

```python
movie_index = movies_data[movies_data.title == close_match]['index'].values[0]
print(movie_index)
```

```python
similarity_score = list(enumerate(similarity[movie_index]))
print(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
print(sorted_similar_movies)
```

```python
print('Movie recommendations: \n')
i = 1
for movie in sorted_similar_movies:
  index = movie[0]
  index_title = movies_data[movies_data.index == index]['title'].values[0]
  if (i<11):
    print(i, '.', index_title)
    i+=1
```

```python

movie_input = input('Enter movie name: ')

list_of_titles = movies_data['title'].tolist()

find_match = difflib.get_close_matches(movie_input, list_of_titles)

close_match = find_match[0]

movie_index = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[movie_index]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True)

print('Movie recommendations: \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  index_title = movies_data[movies_data.index == index]['title'].values[0]
  if (i<11):
    print(i, '.', index_title)
    i+=1
```
